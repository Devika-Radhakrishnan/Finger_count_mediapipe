```python
import cv2
import mediapipe as mp
```

```python
medhands=mp.solutions.hands
draw=mp.solutions.drawing_utils
```

```python
hand=medhands.Hands(max_num_hands=1)
```

```python
video=cv2.VideoCapture(0) #webcam open
while True:
  suc,img=video.read()
  imgrgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  res=hand.process(imgrgb) #no conversion as img and imgrgb are already different no need to convert
  tipids=[4,8,12,16,20] #fingers tip points
  lmlist=[] #landmark list
  cv2.rectangle(img,(20,350),(120,440),(0,255,20),cv2.FILLED)
  if res.multi_hand_landmarks:
     for hand_landmarks in res.multi_hand_landmarks: #for each x and y values
        for id,landmark in enumerate(hand_landmarks.landmark):
          print(id,landmark) 
          cx=landmark.x
          cy=landmark.y
          lmlist.append([id,cx,cy])
          if len(lmlist)!=0 and len(lmlist)==21:
            fingerlist=[]
            #to detect thumb
            if lmlist[20][1]>lmlist[12][1]:
              if lmlist[4][1]>lmlist[3][1]:
                fingerlist.append(0)
              else:
                fingerlist.append(1)
            if lmlist[12][1]>lmlist[20][1]:
               if lmlist[4][1]<lmlist[3][1]:
                 fingerlist.append(0)
               else:
                 fingerlist.append(1)

            for i in range(1,5):
              if lmlist[tipids[i]][2]>lmlist[tipids[i]-2][2]: #to compare the fingertips
                fingerlist.append(0)
              else:
                fingerlist.append(1)
            # print(fingerlist)
            if len(fingerlist)!=0:
              fingercount=fingerlist.count(1)
              print(fingercount)
            cv2.putText(img,str(fingercount),(35,420),cv2.FONT_HERSHEY_COMPLEX_SMALL,3,(0,45,12),4) #fingercount,axis,font,scale,color,thickness

        draw.draw_landmarks(img,hand_landmarks,medhands.HAND_CONNECTIONS,draw.DrawingSpec(color=(255,0,0),thickness=2,circle_radius=2),draw.DrawingSpec(color=(255,255,0),thickness=2)) #draw image, for points
        
  cv2.imshow('HAND',img)
  if cv2.waitKey(1) & 0XFF==ord('q'):
   break
video.release()
cv2.destroyAllWindows()
```
